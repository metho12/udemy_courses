# Visualie Data
  - Start with categories

```python
sns.countplot('is_paid',data=df)
plt.title('pais Vs Free Courses') ## most of the courses is paid
```

```python
df['is_paid'].value_counts()
```

```python
df['is_paid'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(8,5))

sns.countplot('level',data=df) # most of the coures is All Level or Beginner

plt.xlabel('levels')
plt.ylabel('Student Count')
plt.title('Courses Levels Vs Student ')


```

```python
plt.figure(figsize=(8,5))

sns.countplot('subject',data=df) # most of the coures is All Level or Beginner
```
